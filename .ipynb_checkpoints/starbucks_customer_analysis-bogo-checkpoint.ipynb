# Introduction
Periodically, Starbucks sends push notifications to app users, which could be advertisements for a beverage, discount coupons, or a "Buy One Get One" (BOGO) offer. Some customers may not receive any notifications for several weeks. The task involves combining transaction data, demographic data, and push notification data to predict whether a user will be influenced by the app's promotional activities.

The work process detailed involves first preprocessing the data, merging user data with push offer data and transaction data. Then, additional features are added to the user data. Finally, deep learning is used to predict whether a user will be influenced by the app's push notifications, enabling targeted offer dissemination.

# Evaluation Metric
The chosen metric is the 
$$
F_{\beta} = (1 + \beta^2) \cdot \frac{{\text{precision} \cdot \text{recall}}}{{(\beta^2 \cdot \text{precision}) + \text{recall}}}
$$
with β=2), a performance indicator that balances the importance of recall and precision. In this context, recall is given more importance due to the desire to avoid sending irrelevant (or "spammy") push notifications to users. A β value of 2 implies that recall is considered twice as important as precision in this evaluation.

Analysis indicates that 'value' column keys vary by event type: 'offer id' for 'offer received', 'viewed', and 'completed' events, and 'amount' for transactions. We need to split these into separate columns for effective processing

```python
# Convert string representation of dictionary to actual dictionary
df_transcript['value'] = df_transcript['value'].apply(ast.literal_eval)

# Extract 'offer id' and 'amount' into separate columns
def extract_offer_id_and_amount(value_dict):
    offer_id = value_dict.get('offer id') or value_dict.get('offer_id')
    amount = value_dict.get('amount')
    return pd.Series([offer_id, amount], index=['offer id', 'amount'])

df_transcript[['offer id', 'amount']] = df_transcript['value'].apply(extract_offer_id_and_amount)
```

## Data Transformation

### Add more features to the customer profile

For customer profile, to dummfie the gender and member year, and bining the age to group and dummfie it


```python
# Determine the bin edges
min_age = df_customer['age'].min()
max_age = df_customer['age'].max()
bins = range(min_age - min_age % 10, max_age + 10 - max_age % 10, 10)

# Create labels for the bins
labels = [f'{i}-{i + 9}' for i in bins[:-1]]

# Binning the 'age' column
df_customer['age_group'] = pd.cut(df_customer['age'], bins=bins, labels=labels, right=False)

# One-Hot Encoding
age_dummies = pd.get_dummies(df_customer['age_group'], prefix='age')

# Concatenating with original DataFrame
df_customer = pd.concat([df_customer, age_dummies], axis=1)
```

```python
bins
```

```python
labels
```

```python
#Dummfie gender
gender_dummies = pd.get_dummies(df_customer['gender'], prefix='gender_')
df_customer = pd.concat([df_customer, gender_dummies], axis=1)
```

```python
#Dummfie year become member
member_year_dummies = pd.get_dummies(df_customer['member_year'], prefix='ymbr_')
df_customer = pd.concat([df_customer, member_year_dummies], axis=1)
```

```python
#create a feature called tenure to measure their years with starbucks relatively
max_year = df_customer['member_year'].max()
# Calculate the relative tenure
df_customer['relative_tenure'] = max_year - df_customer['member_year'] + 1
```

### Calculating the completion of each offer received by each customer

```python
# Merge the transcript data with the portfolio data to get calculate the expiration date
df_offer_activity=df_transcript[['person','event','time','offer id']][df_transcript['event'] != 'transaction']
```

```python
off_rec = df_offer_merged[df_offer_merged['event'] =='offer received']
off_rec.rename(index = str, columns={"time": "time_rec"}, inplace = True)
off_rec.drop(['event'], inplace = True, axis = 1)
off_rec.head()
```

```python
off_vw = df_offer_merged[df_offer_merged['event'] =='offer viewed']
off_vw.rename(index = str, columns={"time": "time_vw"}, inplace = True)
off_vw.drop(['event'], inplace = True, axis = 1)
off_vw=off_vw[['person',"offer id","time_vw"]]
off_vw.head()
```

```python
off_comp = df_offer_merged[df_offer_merged['event'] =='offer completed']
off_comp.rename(index = str, columns={"time": "time_comp"}, inplace = True)
off_comp.drop(['event'], inplace = True, axis = 1)
off_comp=off_comp[['person',"offer id","time_comp"]]
off_comp.head()
```

```python
#merge the offer view and completion time to offer receive time
df_completion = pd.merge(off_rec, off_vw, on=['person', 'offer id'], how = 'left')
df_completion = pd.merge(df_completion, off_comp, on = ['person', 'offer id'], how = 'left')
```

We calculate the expiration date and also we include all scenarios including users have recived, viewed and completed, or recieved, viewed, and not compelted, recieved not viewed, we only want to consider these trasactions as they are relevant and valid to analysis the offer completion by the customer

```python
df_completion['time_exp'] = df_completion['time_rec'] + df_completion['duration']*24

df_completion= df_completion[((df_completion['time_vw'] >= df_completion['time_rec'])&(df_completion['time_comp'] >= df_completion['time_rec']))\
              |((df_completion['time_vw']>= df_completion['time_rec'])&(df_completion['time_comp'].isnull()))\
              |((df_completion['time_vw'].isnull())&(df_completion['time_comp'].isnull()))]

df_completion.head()
```

```python
df_completion.shape
```

```python
def succ_offer(df):
    if ((df['time_rec']<= df['time_vw'])&\
        (df['time_vw']<= df['time_comp'])&\
        (df['time_comp']<=df['time_exp'])):
        return 1
    else:
        return 0
df_completion['success'] = df_completion.apply(succ_offer, axis = 1)
```

### PCA to reduce dimension

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Separating the target variable 'success' and features
target_bogo_updated = df_BOGO['success']
features_bogo_updated = df_BOGO.drop(['success'], axis=1)

# Filling any NaN values with mean of each column
features_bogo_updated_filled = features_bogo_updated.fillna(0)

# Scaling the features
scaler_bogo_updated = StandardScaler()
features_bogo_updated_scaled = scaler_bogo_updated.fit_transform(features_bogo_updated_filled)

# Applying PCA
pca_bogo_updated = PCA().fit(features_bogo_updated_scaled)

# Re-plotting the Cumulative Summation of the Explained Variance for the updated dataset
plt.figure()
plt.plot(range(1, len(pca_bogo_updated.explained_variance_ratio_) + 1), pca_bogo_updated.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.title('Explained Variance by Different Principal Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

### Data Preparation

```python
# Choosing a suitable number of components based on the variance plot
# A common choice is to choose the number of components that explain 95% of the variance
n_components = sum(pca_bogo_updated.explained_variance_ratio_.cumsum() < 0.95) + 1

# Applying PCA with the chosen number of components
pca_bogo = PCA(n_components=n_components)
X_train_bogo_pca = pca_bogo.fit_transform(features_bogo_updated_scaled)
X_test_bogo_pca = pca_bogo.transform(scaler_bogo_updated.transform(df_BOGO.drop(['success'], axis=1).fillna(0)))

# Splitting the dataset into training and testing sets
X_train_bogo_final, X_test_bogo_final, y_train_bogo_final, y_test_bogo_final = train_test_split(X_train_bogo_pca, target_bogo_updated, test_size=0.2, random_state=42)
```

#### Model 1 : Logistic Regression

```python
# Training the Logistic Regression model
log_reg_bogo = LogisticRegression()
log_reg_bogo.fit(X_train_bogo_final, y_train_bogo_final)

# Evaluating the model
log_reg_bogo_score = log_reg_bogo.score(X_test_bogo_final, y_test_bogo_final)
log_reg_bogo_score
```

Since we focus more on the recall rate, need to test the F2 score

```python
from sklearn.metrics import precision_recall_fscore_support

# Predictions on test set
y_pred_bogo = log_reg_bogo.predict(X_test_bogo_final)

# Calculating precision, recall, and F2 score
precision, recall, f2_score, _ = precision_recall_fscore_support(y_test_bogo_final, y_pred_bogo, beta=2, average='binary')

precision, recall, f2_score
```

#### Model 2 : Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Training the Random Forest model
rf_bogo = RandomForestClassifier(random_state=42)
rf_bogo.fit(X_train_bogo_final, y_train_bogo_final)

# Predictions on test set
y_pred_rf_bogo = rf_bogo.predict(X_test_bogo_final)

# Calculating precision, recall, and F2 score for Random Forest model
precision_rf, recall_rf, f2_score_rf, _ = precision_recall_fscore_support(y_test_bogo_final, y_pred_rf_bogo, beta=2, average='binary')

precision_rf, recall_rf, f2_score_rf
```

#### Model 3 : Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

# Training the Gradient Boosting model
gb_bogo = GradientBoostingClassifier(random_state=42)
gb_bogo.fit(X_train_bogo_final, y_train_bogo_final)

# Predictions on test set
y_pred_gb_bogo = gb_bogo.predict(X_test_bogo_final)

# Calculating precision, recall, and F2 score for Gradient Boosting model
precision_gb, recall_gb, f2_score_gb, _ = precision_recall_fscore_support(y_test_bogo_final, y_pred_gb_bogo, beta=2, average='binary')

precision_gb, recall_gb, f2_score_gb
```

### Gradient Boosting Classifier outperforms the other 2 models, let's do hyperparameter tuning

```python
from sklearn.metrics import make_scorer, fbeta_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier

f2_scorer = make_scorer(fbeta_score, beta=2)

param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

gb = GradientBoostingClassifier(random_state=42)

grid_search = GridSearchCV(gb, param_grid, cv=5, scoring=f2_scorer, verbose=2)

grid_search.fit(X_train_bogo_final, y_train_bogo_final)

print("best_params:", grid_search.best_params_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score

gb_optimal = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
gb_optimal.fit(X_train_bogo_final, y_train_bogo_final)


y_pred_optimal = gb_optimal.predict(X_test_bogo_final)
print("classification report:", classification_report(y_test_bogo_final, y_pred_optimal))

#Feature Importance
feature_importance = gb_optimal.feature_importances_

#Cross Validation
cross_val_scores = cross_val_score(gb_optimal, X_train_bogo_final, y_train_bogo_final, cv=5, scoring='f1')

#Cross Validation Score
print("cross_val_score:", cross_val_scores)

#Feature Importance result
print("feature_importance:", feature_importance)
```
